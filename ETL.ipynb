## VigiMed_Medicamentos



```python
import pandas as pd

# Caminho para o arquivo CSV
caminho_arquivo = '/content/VigiMed_Medicamentos.csv'

# Carregar o arquivo ignorando as linhas problemáticas
df_medicamentos = pd.read_csv(caminho_arquivo, encoding='ISO-8859-1', delimiter=';', on_bad_lines='skip')

# Exibir as primeiras linhas do DataFrame
df_medicamentos.head()

```

Salvar em arquivo Parquet

```python
# Caminho onde o arquivo Parquet será salvo
caminho_parquet = '/content/VigiMed_Medicamentos.parquet'

# Salvar o DataFrame como Parquet
df_medicamentos.to_parquet(caminho_parquet)

# Verificar se o arquivo foi criado
import os
os.path.exists(caminho_parquet)

```

## VigiMed_Notificacoes

```python
import pandas as pd

# Caminho para o arquivo CSV
caminho_arquivo = '/content/VigiMed_Notificacoes.csv'

# Carregar o arquivo ignorando as linhas problemáticas
df_notificacoes = pd.read_csv(caminho_arquivo, encoding='ISO-8859-1', delimiter=';', on_bad_lines='skip')

# Exibir as primeiras linhas do DataFrame
df_notificacoes.head()
```

Salvando em arquivo Parquet

```python
# Converter a coluna PESO_KG para numérico, substituindo valores não numéricos por NaN
df_notificacoes['PESO_KG'] = pd.to_numeric(df_notificacoes['PESO_KG'], errors='coerce')

# Salvar o DataFrame como Parquet
df_notificacoes.to_parquet(caminho_parquet)

```
